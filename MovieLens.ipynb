```python
import pandas as pd
pd.options.display.max_rows=10 #display less 
unames = ['user_id','gender','age','occupation','zip']
users = pd.read_table('Data sets/users.dat',sep='::',header=None,names=unames)
rnames =['user_id','movie_id','rating','timestamp']
ratings = pd.read_table('Data sets/ratings.dat',sep='::',header=None,names=rnames)
mnames=['movie_id','title','genres']
movies= pd.read_table('Data sets/movies.dat',sep='::',header=None, names=mnames)
movies[:5]
```

```python
data = pd.merge(pd.merge(ratings,users),movies)
data
```

```python
# 获得按性别分级的每部电影的评分
mean_ratings = data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250] #选大于250个评分的电影
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
top_female_ratings
```

```python
# F M 评分差异最大的电影
mean_ratings['diff'] = mean_ratings['M']-mean_ratings['F']
sorted_by_diff = mean_ratings.sort_values(by='diff',ascending=False)
sorted_by_diff
```

```python
rating_std_by_title = data.groupby('title')['rating'].std() #争议最大的电影
rating_std_by_title = rating_std_by_title.loc[active_titles]
rating_std_by_title.sort_values(ascending=False)
```
